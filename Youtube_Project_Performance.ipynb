 1. Import Library & Load Data


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV dari file upload
df = pd.read_csv("/content/drive/MyDrive/Youtube_Project/KPI youtube analytics.csv")

# Tampilkan 5 baris awal
df.head(20)
```

2. Pemeriksaan & Pembersihan Data

```python
# Cek Struktur Data

df.info()
```

```python
# Mencari Metrik yang mempunyai data null

df.isnull().sum().sort_values(ascending=False)
```

```python
# Analisis Stastistika Dasar

df.describe().T
```

 3. EDA Awal – Ringkasan KPI

```python
# Melihat data yang paing tinggi menurut Total Views

df.sort_values("total_views", ascending=False).head(5)[["video_id", "total_views", "engagement_rate"]]
```

```python
# Visualisasi Korelasi antar KPI

plt.figure(figsize=(20, 16))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Between KPI Metrics")
```

4. Visualisasi Awal

```python
# Distribusi dari Engagement Rate

plt.figure(figsize=(8, 5))
sns.histplot(df["engagement_rate"], bins=30, kde=True)
plt.title("Distribution of Engagement Rate")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="avg_view_duration_sec", y="engagement_rate", size="total_views", alpha=0.7)
plt.title("Avg View Duration vs Engagement Rate")
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[["engagement_rate", "avg_view_percentage", "net_subscribers"]])
plt.title("Boxplot KPI untuk Outlier Check")
```
